```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
data = pd.read_csv('data/data.csv',index_col=0)
```

```python
columns = data.columns
```

```python
data.index = data.index.astype(str)
```

```python
for column in columns:
    plt.plot(data[column])
    plt.title(column)
    plt.xlabel('year')
    plt.ylabel(column)
    plt.xticks(rotation= 70)
    plt.savefig('README_files/{}_by_year.png'.format(column))
    plt.show();
```

```python
from statsmodels.tsa.stattools import adfuller
```

```python
res = adfuller(data['lyrical mentions percentage'])
```

```python
res
```

```python
res[4]
```

```python
print(f'ADF Statistic: {res[0]}')
print(f'n_lags: {res[1]}')
print(f'p-value: {res[1]}')
for key, value in res[4].items():
    print('Critial Values:')
    print(f'   {key}, {value}')
print('Test conclusion:')
if res[0] < res[4]['1%']:
    print('at the significance level of 1%, the test fails to reject the null hypothesis that there is a unit root')
elif res[0] < res[4]['5%']:
    print('at the significance level of 5%, the test fails to reject the null hypothesis that there is a unit root')
elif res[0] < res[4]['10%']:
    print('at the significance level of 10%, the test fails to reject the null hypothesis that there is a unit root')
else:
    print('the test rejects the null hypothesis. The series is non-stationary.')
```
