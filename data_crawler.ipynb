```python
import pandas as pd
import numpy as np
import re

import lyricsgenius
token = 'zJryEvTUpv-iGO54DHegRpP-ly06kZMFChYJlw5y8AIdPMPZplRyqCLJ7JN2Fi1a'

from collections import Counter

import nltk
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import matplotlib.pyplot as plt
import string
import time
%matplotlib inline
```

```python
rap = pd.read_csv('data/All_Hip_Hop_Songs_from_1958-10-20_to_2018-12-31.csv')
```

```python
rap['year'] = rap['chart_date'].apply(lambda x:x[0:4])
rap.info()
```

```python
rap = rap[['artist','title','year']].drop_duplicates()[rap['year']>='1995']
```

```python
# Include as many stop words as possible
from sklearn.feature_extraction.text import ENGLISH_STOP_WORDS
from gensim.parsing.preprocessing import remove_stopwords, STOPWORDS
import spacy
en = spacy.load('en_core_web_sm')
sw_spacy = en.Defaults.stop_words

STOP = nltk.corpus.stopwords.words('english') + list(string.punctuation) + \
list(ENGLISH_STOP_WORDS) + list(STOPWORDS) + list(sw_spacy)

STOP = set(STOP)
STOP.add('oh')
STOP.add('yeah')
```

```python
lyrics_lst = []
genius = lyricsgenius.Genius(token)
c = 0
for title,artist in zip(rap['title'],rap['artist']):
    try:
        song = genius.search_song(title,artist)
        lyrics = song.lyrics
        lyrics = re.sub(r'\W+', ' ', lyrics)
        lyrics = lyrics.lower()
        lyrics = lyrics.split(' ')
        lyrics = [word for word in lyrics if word not in STOP]
        lyrics = ' '.join(lyrics)
    except:
        lyrics = np.nan
    lyrics_lst.append(lyrics)
    c += 1
    print('{} song(s) finished,5647 in total'.format(c))
```

```python
rap['lyrics'] = lyrics_lst
```
